# Assign BFE to structures


```python
new_gdf = structures_gdf.copy()  # do not change original gdf
coords = structures_gdf.get_coordinates().values

with rasterio.open(f"OpenFileGDB:{geodatabase_path}:{raster_layer_name}") as src:
    # Read the raster data as a numpy array
    raster_data = src.read(1)
    # with raster_data as src:
    new_gdf["BFE"] = [
        x[0] for x in src.sample(coords)
    ]  # this applies the raster depth to the structure point df in a new column, 'BFE'
```

```python
# Structures with no BFE are assigned a value of 1.790000e+308, replace these with 0.0
new_gdf.loc[new_gdf["BFE"] > 1.0e02, "BFE"] = 0
```

# Assign SLOSH depths to structures


```python
list_of_SLOSH_rasters = [
    "us_Category1_SLOSH_Projected",
    "us_Category2_SLOSH_Projected",
    "us_Category3_SLOSH_Projected",
    "us_Category4_SLOSH_Projected",
    "us_Category5_SLOSH_Projected",
]
```

```python
# check raster coords

raster_layer_name = "us_Category1_SLOSH_Projected"

# Open the raster dataset
with rasterio.open(f"OpenFileGDB:{geodatabase_path}:{raster_layer_name}") as src:
    # Read the raster data as a numpy array
    raster_data = src.read(1)

    # Access metadata
    print(src.crs)  # Coordinate reference system
    print(src.bounds)  # Bounding box
    print(src.width, src.height)  # Dimensions
```

```python
coords = structures_gdf.get_coordinates().values

for category in list_of_SLOSH_rasters:
    with rasterio.open(f"OpenFileGDB:{geodatabase_path}:{category}") as src:
        # Read the raster data as a numpy array
        raster_data = src.read(1)
        # with raster_data as src:
        new_gdf[f"{category}_depth"] = [
            x[0] for x in src.sample(coords)
        ]  # this applies the raster depth to the structure point df in a new column, named for the storm category
```

```python
new_gdf
```

```python
new_gdf.value_counts("us_Category1_SLOSH_Projected_depth")
```

```python
# Structures with no SLOSH depths are assigned a value of 255, replace these with 0.0
for category in list_of_SLOSH_rasters:
    new_gdf.loc[new_gdf[f"{category}_depth"] > 250, f"{category}_depth"] = 0
```

# Process data for simulation


```python
import ast

full_structures_data = pd.read_csv("data/full_structures_data_all_flood_depths.csv")

# Convert string representations of lists back to actual lists
full_structures_data["SLOSH_depths"] = full_structures_data["SLOSH_depths"].apply(
    ast.literal_eval
)
full_structures_data["flood_elevation_list"] = full_structures_data["SLOSH_depths"]

# Filter out rows where flood_elevation_list contains 99
full_structures_data = full_structures_data[
    ~full_structures_data["flood_elevation_list"].apply(lambda x: 99 in x)
]

# Calculate 1% of the total rows
# num_rows_to_select = int(len(full_structures_data) * 0.01)

# Randomly select 1% of the rows
# simulation_df = full_structures_data.sample(n=num_rows_to_select)

simulation_df = full_structures_data.copy()

# add 2022 fips to the structure data
from shapely.geometry import Point

simulation_df["geometry"] = simulation_df.apply(
    lambda row: Point(row["x"], row["y"]), axis=1
)  # Convert to GeoDataFrame
simulation_df = gpd.GeoDataFrame(
    simulation_df, geometry="geometry"
)  # Set CRS to WGS84 (EPSG:4326)
simulation_df.crs = "EPSG:2278"

year = 2020
county_shp = gpd.read_file(
    "https://www2.census.gov/geo/tiger/TIGER{}/TRACT/tl_{}_48_tract.zip".format(
        year, year
    )
)
# county_shp.plot()
county_shp = county_shp[["GEOID", "geometry"]]

county_shp = county_shp.to_crs("EPSG:2278")
print(county_shp.head())



# Display the sampled DataFrame
joint_df = gpd.sjoin(simulation_df, county_shp, how="left", predicate="within")
joint_df.head()
print(joint_df["GEOID"].nunique())
joint_df.to_csv("data/full_data_for_simulation.csv", index=False)
```
