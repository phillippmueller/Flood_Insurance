```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
tracks = gpd.read_file('data\hurricane_track_data\IBTrACS.NA.list.v04r01.lines.shp')
```

```python
tracks
```

```python
tracks.columns
```

```python
cols_to_keep = ['SID','SEASON', 'NUMBER', 'SUBBASIN', 'NAME', 'ISO_TIME', 'USA_WIND', 'USA_STATUS', 'USA_SSHS',
                'year', 'month', 'day', 'hour', 'min', 'geometry']
```

```python
cleantracks = tracks[cols_to_keep]
```

```python
cleantracks
```

```python
cleantracks = cleantracks[(cleantracks['SEASON'] >= 1901)]
```

```python
cleantracks
```

```python
# assign a value fo 0 to all negative SSHS values
cleantracks['SSHS_clean'] = cleantracks['USA_SSHS'].apply(lambda x: 0 if x <= 0 else x)
```

```python
cleantracks
```

```python
cleantracks.to_file('data\hurricane_track_data\IBTrACS_NA_20240819_SSHSclean_1901.shp')  
```
