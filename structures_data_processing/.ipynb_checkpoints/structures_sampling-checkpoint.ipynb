# Info

This notebook samples National Structure Inventory (NSI) data from the US Army Corps of Engineers (USACE) for use in flood risk assessment modeling. The data corresponds to Galveston County, Texas. The `cbfips` field can be used to merge the structures sample data with US Census data and others.

Note: it is recommended to create a virtual environment (conda or venv) to run geopandas

# Import data

GIS data was processed externally using ArcGIS Pro to clip files to the desired extents and export layers to a geodatabase, `flood_risk_assessment.gdb`.

Links to data information:

- NSI data dictionary: https://www.hec.usace.army.mil/confluence/nsi/technicalreferences/2022/technical-documentation
- FEMA flood zone information: https://www.fema.gov/glossary/flood-zones
- Social Vulnerability Index: https://www.atsdr.cdc.gov/placeandhealth/svi/index.html


## Import Census data


```python
# Set API key
c = Census("0f127583bd5f4f250742b6b83712defba43beeea")

TX_census = c.acs5.state_county_tract(
    fields=[
        "NAME",  # Geographic name (tract name)
        "B19013_001E",  # Median Household Income
        # "B19301_001E",  # Per Capita Income
        "B02001_001E",  # Total Population
        "B02001_002E",  # White Alone
        "B02001_003E",  # Black or African American Alone
        "B02001_005E",  # Asian Alone
        "B02001_004E",  # American Indian and Alaska Native Alone
        "B03001_003E",  # Hispanic or Latino Origin
        "B25081_002E",  # Housing Units with a Mortgage
        "B25081_003E",  # Housing Units without a Mortgage
        # "B25088_002E",  # Median Monthly Owner Costs (with mortgage)
        # "B25088_003E",  # Median Monthly Owner Costs (without mortgage)
        "B15003_001E",  # Educational Attainment for Population 25 Years and Over
        "B15003_022E",  # Bachelor's Degree
        "B15003_023E",  # Master’s Degree
        "B15003_024E",  # Professional School Degree
        "B15003_025E",  # Doctorate Degree
        # "B25077_001E",  # Median Value of Owner-Occupied Housing Units
        "B25003_002E",  # Owner-Occupied Housing Units
        "B25003_003E",  # Renter-Occupied Housing Units
        # "C17002_001E",  # Total Population for whom poverty status is determined
        # "C17002_002E",  # Population below poverty level (under 50% of poverty threshold)
        # "C17002_003E",  # Population between 50% and below poverty level
        "B01003_001E",  # Total Population
    ],
    state_fips=states.TX.fips,  # FIPS code for Texas
    county_fips="*",  # All counties
    tract="*",  # All tracts within counties
    year=2022,  # Specifying the ACS year
)
```

```python
census_df = pd.DataFrame(TX_census)
census_df = census_df.rename(
    columns={
        "B19013_001E": "Median_Household_Income",
        # "B19301_001E": "Per_Capita_Income",
        "B02001_001E": "Total_Population",
        "B02001_002E": "White_Alone",
        "B02001_003E": "Black_Alone",
        "B02001_005E": "Asian_Alone",
        "B02001_004E": "Native_American_Alone",
        "B03001_003E": "Hispanic_Origin",
        "B25081_002E": "Housing_Units_With_Mortgage",
        "B25081_003E": "Housing_Units_Without_Mortgage",
        # "B25088_002E": "Median_Monthly_Owner_Cost_With_Mortgage",
        # "B25088_003E": "Median_Monthly_Owner_Cost_Without_Mortgage",
        "B15003_001E": "Educational_Attainment_Total",
        "B15003_022E": "Bachelors_Degree",
        "B15003_023E": "Masters_Degree",
        "B15003_024E": "Professional_Degree",
        "B15003_025E": "Doctorate_Degree",
        # "B25077_001E": "Median_Value_Owner_Occupied_Housing",
        "B25003_002E": "Owner_Occupied_Units",
        "B25003_003E": "Renter_Occupied_Units",
        # "C17002_001E": "Total_Poverty_Status_Population",
        # "C17002_002E": "Population_Below_50_Poverty_Level",
        # "C17002_003E": "Population_50_to_Poverty_Level",
        "B01003_001E": "Total_Population_Count",
    }
)
# Show the dataframe
print(census_df.head(2))
print("Shape: ", census_df.shape)
```

## Import Structure layers


```python
NSI_GalvestonCo = gpd.read_file(
    "flood_risk_assessment.gdb", driver="FileGDB", layer="Galveston_Co_NSI"
)
```

```python
NSI_GalvestonCo
```

```python
NSI_GalvestonCo.columns
```

```python
# Extract relevant columns; adjust this list to keep data columns as desired
cols_to_keep = [
    "fd_id",
    "firmzone",
    "ground_elv",
    "sqft",
    "st_damcat",
    "val_struct",
    "cbfips",
    "x",
    "y",
    "geometry",
]

NSI_GalvestonCo = NSI_GalvestonCo[cols_to_keep]
NSI_GalvestonCo
```

```python
NSI_GalvestonCo_res = NSI_GalvestonCo[(NSI_GalvestonCo["st_damcat"] == "RES")]
NSI_GalvestonCo_res
```

```python
NSI_GalvestonCo_res["firmzone"].value_counts()
```

```python
print(
    "Percentage of Galveston Co residential structures in a FEMA flood zone:",
    round(
        (
            NSI_GalvestonCo_res["firmzone"].value_counts().sum()
            / len(NSI_GalvestonCo_res)
        )
        * 100,
        2,
    ),
    "%",
)
```

```python
NSI_GalvestonCo_res_zone = NSI_GalvestonCo_res.dropna()
NSI_GalvestonCo_res_zone
```

```python
NSI_GalvestonCo_res_zone["firmzone"].value_counts()
```

```python
NSI_GalvestonCo_res_zone.plot()
```

```python
NSI_GalvestonCo_res_zone.crs
```

# Plot NSI data with FEMA hazard zones


```python
Texas_counties = gpd.read_file(
    "flood_risk_assessment.gdb", driver="FileGDB", layer="Texas_counties"
)
```

```python
Texas_counties.crs
```

```python
Texas_counties.plot()
```

```python
# convert NSI data to proper coordinate reference system (CRS) for plotting
NSI_GalvestonCo_res_zone = NSI_GalvestonCo_res_zone.to_crs(Texas_counties.crs)
```

```python
# Plot all residential FEMA zone Galveston NSI data

fig, ax = plt.subplots()
Texas_counties.plot(ax=ax, color="forestgreen", alpha=0.25)
# FEMA_zones.plot(ax=ax, column='FLD_ZONE', cmap='viridis', alpha=0.5)
NSI_GalvestonCo_res_zone.plot(ax=ax, column="firmzone", cmap="viridis", markersize=0.1)

minx, miny, maxx, maxy = NSI_GalvestonCo_res_zone.geometry.total_bounds
ax.set_xlim((minx - 0.025, maxx + 0.025))
ax.set_ylim((miny - 0.025, maxy + 0.025))

ax.set_axis_off()

# plt.savefig('../Galveston_NSI.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Plot all residential FEMA zone Galveston NSI data against FEMA zones

fig, ax = plt.subplots()
Texas_counties.plot(ax=ax, color="forestgreen", alpha=0.25)
FEMA_zones.plot(ax=ax, column="FLD_ZONE", cmap="viridis", alpha=0.25)
NSI_GalvestonCo_res_zone.plot(ax=ax, column="firmzone", cmap="viridis", markersize=0.1)

minx, miny, maxx, maxy = NSI_GalvestonCo_res_zone.geometry.total_bounds
ax.set_xlim((minx - 0.025, maxx + 0.025))
ax.set_ylim((miny - 0.025, maxy + 0.025))

# ax.set_axis_off()

# plt.savefig('../Galveston_NIS_FEMA_zones.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Plot zoomed-in residential FEMA zone Galveston NSI data against FEMA zones

fig, ax = plt.subplots()
Texas_counties.plot(ax=ax, color="forestgreen", alpha=0.25)
FEMA_zones.plot(ax=ax, column="FLD_ZONE", cmap="viridis", alpha=0.25)
NSI_GalvestonCo_res_zone.plot(ax=ax, column="firmzone", cmap="viridis", markersize=0.1)

minx, miny, maxx, maxy = NSI_GalvestonCo_res_zone.geometry.total_bounds
ax.set_xlim((-10575000, -10560000))
ax.set_ylim((3400000, 3410000))

# ax.set_axis_off()

# plt.savefig('../Houston-Galveston_FEMA_zones_ZOOMED.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Plot zoomed-in residential FEMA zone Galveston NSI data against FEMA zones

fig, ax = plt.subplots()
Texas_counties.plot(ax=ax, color="forestgreen", alpha=0.25)
FEMA_zones.plot(ax=ax, column="FLD_ZONE", cmap="viridis", alpha=0.25)
NSI_GalvestonCo_res_zone.plot(ax=ax, column="firmzone", cmap="viridis", markersize=0.1)

minx, miny, maxx, maxy = NSI_GalvestonCo_res_zone.geometry.total_bounds
ax.set_xlim((-10588000, -10576000))
ax.set_ylim((3430000, 3440000))

# ax.set_axis_off()

# plt.savefig('../Houston-Galveston_FEMA_zones_ZOOMED2.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```

# Sample structures and export

Sample $n$ structures for simulation input. There are 43,858 residential structure in a FEMA flood zone in Galveston County from which we may sample.

Additional processing may be accomplished using the `cbfips` field to merge the structures data with Census block/tract data or SVI data (provided it also has a census unit identifier field).


```python
# change this input value to adjust the sample size

structures_sample = NSI_GalvestonCo_res_zone.sample(1000)
structures_sample
```

```python
# Visualize sample data

fig, ax = plt.subplots()
Texas_counties.plot(ax=ax, color="forestgreen", alpha=0.25)
FEMA_zones.plot(ax=ax, color="steelblue", alpha=0.25)
structures_sample.plot(ax=ax, color="red", markersize=0.1)

minx, miny, maxx, maxy = FEMA_zones.geometry.total_bounds
ax.set_xlim((minx - 0.025, maxx + 0.025))
ax.set_ylim((miny - 0.025, maxy + 0.025))

ax.set_axis_off()

# plt.savefig('../structures_sample_map.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
structures_sample.drop("geometry", axis=1).to_csv(
    "residential_structures.csv", index=False
)
```
